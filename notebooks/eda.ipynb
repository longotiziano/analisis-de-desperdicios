```python

import sys, os

# Obtengo la ruta ra铆z del proyecto (subo un nivel desde /notebooks)
project_root = os.path.abspath("../")

# La agrego al path si no est谩
if project_root not in sys.path:
    sys.path.append(project_root)

# Importado de librer铆as
from scripts.utils import realizar_request, detectar_outliers_iqr
import pandas as pd
import os
import json
import subprocess
import tabula
```

## 2. Obtenci贸n de datasets

- Obtenci贸n del dataset de ventas y paso a DataFrame

```python
# Cargar credenciales de Kaggle desde un archivo JSON
with open(r"..\kaggle_keys.json") as f:
    kaggle_creds = json.load(f)

os.environ["KAGGLE_USERNAME"] = kaggle_creds["username"]
os.environ["KAGGLE_KEY"] = kaggle_creds["key"]

dataset = "virtualschool/restaurant-cost-and-sales-dataset"

# Ejecuta el comando kaggle datasets download
subprocess.run([
    "kaggle", "datasets", "download",
    "-d", dataset,
    "--unzip",
    "-p", '../data/raw'
])

dir_raw = r"..\data\raw"
excel_restaurante = dir_raw + r"\Restaurant_Data.xlsx"

df_orders = pd.read_excel(excel_restaurante, sheet_name="Orders")
df_items = pd.read_excel(excel_restaurante, sheet_name="Items")
```

- Dataset del INDEC

```python
path_informe = dir_raw + r"\INFORME_INDEC_1-10-25.pdf"

# Obtener DataFrame que me interesa
df_indec = tabula.read_pdf(
    path_informe,
    pages=12,
    multiple_tables=True,
    encoding='latin1'  
)[0]
```

- Valor del d贸lar

```python
url_dolarapi = 'https://dolarapi.com/v1/dolares/blue'

request_ok, data = realizar_request(url_dolarapi)
if not request_ok:
    raise RuntimeError("No se pudo obtener el valor del d贸lar desde la API.")
dolar_blue = float(data.get("venta"))
```

## 3. Analizado de DataFrames y detecci贸n de tareas

- Ahora, **detecci贸n de nulos y duplicados**:

```python
dfs = {
    "df_orders": df_orders,
    "df_items": df_items,
    "df_indec": df_indec
}

for nombre, df in dfs.items():
    print(f"-> {nombre} - {df.isnull().sum().sum()} valores nulos --- {df.duplicated().sum().sum()} valores duplicados")
```

- B煤squeda de outliers

```python
for nombre, df in dfs.items():
    print(f"\n Analizando: {nombre}")
    
    # Tomar solo columnas num茅ricas
    cols_numericas = df.select_dtypes(include="number").columns
    
    for col in cols_numericas:
        outliers = detectar_outliers_iqr(df, col)
        print(f"  - Columna {col}: {len(outliers)} outliers")
```
